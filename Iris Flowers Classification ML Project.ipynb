```python
#import libraries 

import numpy as np                         
import pandas as pd                         
import seaborn as sns                        
import matplotlib.pyplot as plt              
%matplotlib inline

sns.set()
```

## load dataset

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
ax = iris.plot(figsize=(15,8), title='Iris Dataset')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
```

# performing EDA 

```python
df = iris.drop(['species'], axis = 1)
```

### first row 

```python
df.iloc[0]
```

```python
df.iloc[0].plot(kind='bar')
```

### Histogram

```python
iris.plot.hist()
```

```python
iris.plot(kind = 'hist', stacked = False, bins = 20)
```

```python
iris.plot(kind = 'hist', stacked = True, bins = 10, orientation = 'horizontal')
```

### first discrete difference

```python
iris['sepal_width'].diff()
```

```python
iris['sepal_width'].diff().plot(kind = 'hist', stacked = True, bins = 20)
```

```python
df.diff().head()
```

```python
df.diff().hist(color = 'b', alpha = 0.8, figsize=(8,8))
```

### Scatter Plots 

```python
df.plot.scatter(color='b' , x = 'sepal_length', y = 'petal_length')
```

```python
df.plot.scatter(x = 'sepal_length', y = 'petal_length', c = 'sepal_width')
```
